```python
from tkinter import *
from  tkinter import  messagebox
from openpyxl import load_workbook
from datetime import datetime
from tkinter import ttk

class BMI_Class:
    
 
    
    def __init__(self,root):
        self.root=root
        self.root.title("BMI System")
        self.root.geometry("1300x750+140+30")
        self.root.config(bg="#E0FFFF")
        self.root.resizable(width=False, height=False)

        title = Label(self.root, text="BMI Program",font =("Elephant",25),bg="#E0FFFF").pack(side=TOP,fill=X)
        title_bottom = Label(self.root, text="Chào mừng lớp PYF2206R1",font =("times new roman",15),bg="gray").pack(side=BOTTOM,fill=X)

#          ======= Variable =========
        self.var_Identify=StringVar()
        self.var_height=StringVar()
        self.var_weight=StringVar()
        self.var_yourBMI=StringVar()
        self.var_status=StringVar()
        self.var_level=StringVar()
        lbl_Identify=Label(self.root,text="Identify",font=("goudy old style",20),bg="#E0FFFF").place(x=700,y=100)
        txt_Identify=Entry(self.root,textvariable=self.var_Identify,font=("times new roman",15),bg="lightyellow").place(x=850,y=100,width=300,height=40)
        
        lbl_height=Label(self.root,text="Enter height",font=("goudy old style",20),bg="#E0FFFF").place(x=50,y=100)
        txt_height=Entry(self.root,textvariable=self.var_height,font=("times new roman",15),bg="lightyellow").place(x=250,y=100,width=300,height=40)

        lbl_weight=Label(self.root,text="Enter weight",font=("goudy old style",20),bg="#E0FFFF").place(x=50,y=170)
        txt_weight=Entry(self.root,textvariable=self.var_weight,font=("times new roman",15),bg="lightyellow").place(x=250,y=170,width=300,height=40)

        btn_BMI=Button(self.root,text="Caculate BMI",bg="#47d1d1",command=self.BMI,font=("times new roman",15)).place(x=250,y=240,width=130,height=50)
        btn_clear=Button(self.root,text="Clear",bg="gray",command=self.clear,font=("times new roman",15)).place(x=420,y=240,width=130,height=50)

        lbl_your_BMI = Label(self.root, text="Your BMI", font=("goudy old style", 20), bg="#E0FFFF").place(x=50, y=320)
        txt_your_BMI = Entry(self.root,textvariable=self.var_yourBMI, font=("times new roman", 20), bg="lightyellow",fg="#cc2900",state="readonly").place(x=250, y=320, width=300, height=40)

        lbl_state = Label(self.root, text=" Your status", font=("goudy old style", 20), bg="#E0FFFF").place(x=50, y=400)
        txt_state = Entry(self.root, textvariable=self.var_status, font=("times new roman", 20), bg="lightyellow",fg="#cc2900",state="readonly").place(x=250, y=400, width=300, height=40)

        lbl_weight = Label(self.root, text="Disease level", font=("goudy old style", 20), bg="#E0FFFF").place(x=50, y=480)
        txt_weight = Entry(self.root, textvariable=self.var_level, font=("times new roman", 20), bg="lightyellow",fg="#cc2900",state="readonly").place(x=250, y=480, width=300, height=40)
        btn_exit=Button(self.root,text="Exit",bg="gray",command=self.exit,font=("times new roman",15)).place(x=300,y=540,width=130,height=50)
        
        btn_saveBMI=Button(self.root,text="Save",bg="#47d1d1",command=self.saveBMI,font=("times new roman",15)).place(x=80,y=240,width=130,height=50)
        
        
    def clear(self):
        self.var_Identify.set("")
        self.var_height.set("")
        self.var_weight.set("")
        self.var_yourBMI.set("")
        self.var_status.set("")
        self.var_level.set("")
        self.clear_all()  

    def clear_all(self):
        for i in tv.get_children():
            tv.delete(i)
              
    def show(self):
        
#         self.vsb = ttk.Scrollbar(self.root, orient="horizontal", command=tv.xview)
#         self.vsb.place(x=650+200, y=,width=300,height=20 )
#         tv.config(xscrollcommand=self.vsb.set)
#         self.vsb.config(command=tv.xview)
        
 
          
        tv['columns']=('Identify', 'Height', 'Weight','Your BMI','Your Status','Disease level','Time')
        tv.column('#0', width=0, stretch=NO)
        tv.column('Identify', anchor=CENTER, width=80)
        tv.column('Height', anchor=CENTER, width=80)
        tv.column('Weight', anchor=CENTER, width=80)
        tv.column('Your BMI', anchor=CENTER, width=80)
        tv.column('Your Status', anchor=CENTER, width=80)
        tv.column('Disease level', anchor=CENTER, width=80)  
        tv.column('Time', anchor=CENTER, width=120)

        tv.heading('#0', text='', anchor=CENTER)
        tv.heading('Identify', text='Identify', anchor=CENTER)
        tv.heading('Height', text='Height', anchor=CENTER)
        tv.heading('Weight', text='Weight', anchor=CENTER)
        tv.heading('Your BMI', text='Your BMI', anchor=CENTER)
        tv.heading('Your Status', text='Your Status', anchor=CENTER)
        tv.heading('Disease level', text='Disease level', anchor=CENTER)
        tv.heading('Time', text='Time', anchor=CENTER)
        wb = load_workbook('BMI-1.xlsx')
        ws = wb.active
        self.clear_all()          
        for row_num in range(1, ws.max_row):
            if(ws[f'A{row_num+1}'].value == self.var_Identify.get()):
                tv.insert(parent='', index=(row_num-1), iid=(row_num-1), text='', values=(ws[f'A{row_num+1}'].value,
                                                                                          ws[f'B{row_num+1}'].value,
                                                                                          ws[f'C{row_num+1}'].value,
                                                                                          ws[f'D{row_num+1}'].value,
                                                                                          ws[f'E{row_num+1}'].value,
                                                                                         ws[f'F{row_num+1}'].value,
                                                                                         ws[f'G{row_num+1}'].value))
        
            
           
  
        tv.place(x=650, y=200)
    

    def saveBMI(self):
        self.show()
               
    def BMI(self):
        if self.var_weight.get()=="" or self.var_height.get()=="" or  self.var_Identify.get()=="":
            messagebox.showerror("Error","Please enter your height and weight and identify")
        else:
            self.h=float(self.var_height.get())
            self.w=float(self.var_weight.get())
            self.number=round(self.w/(self.h*self.h))

            if self.number <=18.5:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Thin"
                self.var_status.set(self.var_status.value)
                self.var_level.value = "Low"
                self.var_level.set(self.var_status.value)
                

            elif self.number <=24.9:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Normal"
                self.var_status.set(self.var_status.value)
                self.var_level.value = "Medium"
                self.var_level.set(self.var_status.value)

            elif self.number <=29.9:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Lightly Overweight"
# Thừa cân nhẹ 
                self.var_status.set(self.var_status.value)
                self.var_level.value = "A little high"
# Hơi nặng
                self.var_level.set(self.var_status.value)

            elif self.number <=34.9:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Obesity level 1"
# Béo phì cấp độ 1
                self.var_status.set(self.var_status.value)
                self.var_level.value = "High"
                self.var_level.set(self.var_status.value)

            elif self.number <= 39.9:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Obesity level 2"
# Béo phì cấp độ 2
                self.var_status.set(self.var_status.value)
                self.var_level.value = "Very high"
                self.var_level.set(self.var_status.value)
            else:
                self.var_yourBMI.value = self.number
                self.var_yourBMI.set(self.number)
                self.var_status.value = "Obesity level 3"
# Béo phì cấp độ 3
                self.var_status.set(self.var_status.value)
                self.var_level.value = "Dangerous"
                self.var_level.set(self.var_status.value)
            now = datetime.now()
            dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
            wb = load_workbook('BMI-1.xlsx')
            ws = wb.active 
            print(ws.cell(row=1, column=1).value)
            row = ws.max_row+1
            ws.cell(row,1).value = self.var_Identify.get()
            ws.cell(row,2).value = self.h
            ws.cell(row,3).value = self.w
            ws.cell(row,4).value = self.var_yourBMI.value
            ws.cell(row,5).value = self.var_status.value
            ws.cell(row,6).value = self.var_level.value
            ws.cell(row,7).value = dt_string
            wb.save('BMI-1.xlsx')   
            
    def exit(self):
        op=messagebox.askyesno("Confirm","Do you readly want to exit")
        if op==True:
            self.root.destroy()
        

if __name__=="__main__":
    root = Tk()
    obj=BMI_Class(root)
    tv = ttk.Treeview(root)
    root.mainloop()


```
